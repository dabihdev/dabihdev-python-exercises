## 1. Theoretical Overview & Syntax Guide: The `datetime` Library

Python provides the built-in `datetime` module to handle dates, times, and time intervals. Below are the core objects and methods you need to know.

### Core Objects
* `datetime.date`: Represents a date (Year, Month, Day) without time. Example: `datetime.date(2026, 6, 13)`
* `datetime.time`: Represents time (Hour, Minute, Second, Microsecond) independent of any date. Example: `datetime.time(14, 30, 0)`
* `datetime.datetime`: Combines both date and time attributes. Example: `datetime.datetime(2026, 6, 13, 14, 30, 0)`
* `datetime.timedelta`: A duration representing the difference between two date or datetime objects down to microsecond resolution.

### Essential Syntax and Methods

#### 1. Getting Current Datetime
```python
import datetime

now = datetime.datetime.now()        # Current local date and time
today = datetime.date.today()       # Current date only
```

#### 2. Parsing Strings into Datetime Objects (`strptime`)
When reading dates from raw texts or files, you often need to convert strings to datetime objects using `datetime.strptime(string, format)`.
```python
date_string = "2026-06-13 14:30:00"
parsed_date = datetime.datetime.strptime(date_string, "%Y-%m-%d %H:%M:%S")
```

#### 3. Formatting Datetime Objects into Strings (`strftime`)
To output or display dates in specific formats, use `datetime_obj.strftime(format)`.
```python
formatted_string = parsed_date.strftime("%A, %B %d, %Y")
print(formatted_string)  # Output: Saturday, June 13, 2026
```

#### Common Format Codes:
* `%Y`: 4-digit Year (e.g., 2026)
* `%m`: 2-digit Month (01 to 12)
* `%d`: 2-digit Day of the month (01 to 31)
* `%H`: Hour in 24-hour format (00 to 23)
* `%M`: Minute (00 to 59)
* `%S`: Second (00 to 59)
* `%A`: Full weekday name (e.g., Saturday)
* `%B`: Full month name (e.g., June)

#### 4. Date Arithmetic with `timedelta`
You can shift dates forward or backward by adding or subtracting a `timedelta` object.
```python
from datetime import datetime, timedelta

today = datetime.now()
ten_days_later = today + timedelta(days=10)
three_weeks_ago = today - timedelta(weeks=3)

# Calculating differences yields a timedelta object
duration = ten_days_later - today
print(duration.days)  # Output: 10
```
