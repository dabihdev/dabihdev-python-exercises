# Esercizi su Istruzioni If-Else

In questi esercizi dovrai usare la struttura:
```python

    # cosa fare se vero
else:
    # cosa fare se falso
```

````python
# Esercizio 0: Esempio di soluzione
# Crea una variabile 'numero' con valore 10.
# Se il numero è maggiore di 0, stampa "Positivo", altrimenti stampa "Negativo o zero".

numero = 10
if numero > 0:
    print("Positivo")
else:
    print("Negativo o zero")
````
