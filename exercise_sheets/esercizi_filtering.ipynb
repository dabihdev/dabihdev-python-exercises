# 🐍 Esercizi: Filtrare le Liste (FOR + IF)

Benvenuto in questo Jupyter Notebook dedicato al filtraggio delle liste in Python. Il filtraggio è un'operazione fondamentale in programmazione che consiste nell'analizzare una collezione di dati (lista di partenza) e selezionare solo gli elementi che soddisfano un determinato criterio.

### 💡 Il Pattern di Risoluzione Comune

Per risolvere quasi tutti gli esercizi di questo notebook, seguirai uno schema standard in 5 passaggi:

1. **Lista di Partenza:** Hai già una lista popolata con dei dati originari (es. `numeri = [...]`).
2. **Inizializzazione della Lista Vuota:** Crea una nuova variabile associata a una lista vuota (es. `risultato = []`). Servirà a contenere gli elementi filtrati.
3. **Ciclo `for`:** Itera su ogni elemento della lista originale per esaminarlo uno alla volta (`for elemento in lista_partenza:`).
4. **Controllo Condizionale `if`:** All'interno del ciclo, inserisci una condizione booleana (`if condizione:`).
5. **Metodo `.append()`:** Se la condizione del punto 4 è vera (`True`), aggiungi l'elemento corrente alla nuova lista usando `risultato.append(elemento)`.
