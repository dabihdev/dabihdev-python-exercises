## Spiegazione: Come funziona un contatore?

Un contatore è una semplice variabile numerica (di solito inizializzata a `0`) che viene utilizzata per tenere traccia di quante volte si verifica un determinato evento mentre scorriamo gli elementi di una collezione (come una lista).

La struttura standard di uno script con contatore segue sempre **5 passi fondamentali**:

1. **Inizializzazione**: Si crea la variabile del contatore (es. `contatore = 0`) *prima* che il ciclo abbia inizio.
2. **Il Ciclo**: Si usa un ciclo `for` per esaminare uno alla volta tutti gli elementi della lista.
3. **La Condizione**: All'interno del ciclo, si inserisce un blocco `if` per verificare se l'elemento corrente soddisfa il requisito richiesto dalla traccia.
4. **L'Incremento**: Se la condizione dell'effetto `if` è vera (`True`), si aumenta il valore del contatore di 1 usando l'operatore compatto `+= 1`.
5. **La Stampa**: Una volta che il ciclo ha terminato di scorrere tutta la lista, si stampa il valore finale accumulato.

---

### Esempio Guida (Esercizio 0)
**Traccia:** Data una lista di numeri, calcola quanti di questi sono maggiori di 5 e stampa il risultato finale.

```python
# Lista di partenza per il test
lista_numeri = [2, 7, 4, 9, 10]

# 1. Inizializzo il contatore a zero
contatore = 0  

# 2. Ciclo for per scorrere la lista
for numero in lista_numeri:
    # 3. Condizione if per filtrare i numeri
    if numero > 5:
        # 4. Incremento del contatore
        contatore += 1  

# 5. Stampa del risultato finale (In questo caso stamperà 3)
print(contatore)
```
