## Topic 1: Rows and Column Selection

### Introduction
In pandas, selecting specific slices of data is a fundamental skill. 
- **Column Selection**: You can select a single column using `df['column_name']` or multiple columns using a list of names: `df[['col_A', 'col_B']]`.
- **Label-based Selection (`.loc`)**: Used to select rows and columns by their text labels or index values: `df.loc[row_label, col_label]`.
- **Integer-based Selection (`.iloc`)**: Used to select rows and columns by their zero-indexed positions: `df.iloc[row_index, col_index]`.

### Example
To select the first 5 rows of the `date` and `mean_temp` columns using integer positioning:

## Topic 2: Conditional Selection

### Introduction
Conditional selection filters a dataframe to show only rows where a specific condition is met. We pass a boolean expression inside brackets: `df[condition]`.
When filtering with multiple criteria, we combine them using bitwise logical operators:
- `&` (AND): Both conditions must evaluate to `True`.
- `|` (OR): At least one condition must evaluate to `True`.

**Crucial Rule:** Every individual condition must be enclosed in parentheses `( )` due to operator precedence rules in Python.

### Example
To select rows where the `mean_temp` is greater than 20°C and `precipitation` is 0:
```python
dry_warm_days = df[(df['mean_temp'] > 20) & (df['precipitation'] == 0)]
```

## Topic 3: Aggregating Methods

### Introduction
Aggregation reduces multiple data points down to a single summary metric. Pandas comes with built-in statistical functions that ignore missing values (`NaN`) by default. 
- `df['col'].mean()`: Returns the average value.
- `df['col'].min()`: Returns the minimum value.
- `df['col'].max()`: Returns the maximum value.
- `df['col'].sum()`: Returns the sum total.

### Example
Finding the highest global radiation recorded:
```python
highest_radiation = df['global_radiation'].max()
print(f"Max radiation recorded: {highest_radiation}")
```
