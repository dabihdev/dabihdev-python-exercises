## Theory: What is GroupBy?
Grouping allows you to split your data into buckets based on a category (like the car company) and then calculate a statistic (like the average price, total count, or maximum value) for each bucket.

### Basic Syntax Template:
```python
# Basic formula:
df.groupby('Category_Column')['Numeric_Column'].statistic()
```

### Common Statistics you can use:
- `.mean()` : Calculates the average
- `.sum()` : Adds all values together
- `.count()` : Counts the number of items
- `.max()` : Finds the highest value
- `.min()` : Finds the lowest value
