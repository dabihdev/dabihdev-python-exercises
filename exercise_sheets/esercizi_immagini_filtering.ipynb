# Esercizi Python: Ricerca Lineare nelle Matrici di Pixel

In questo notebook ci concentreremo sul pattern fondamentale per esplorare le immagini: la **ricerca lineare tramite cicli annidati**.

Quando lavoriamo con le immagini (matrici di pixel), il modo più robusto e standard per navigare al loro interno consiste nel calcolare prima le dimensioni della matrice (altezza e larghezza) e poi usare due cicli `for ... in range(...)` sovrapposti.

---

### La Struttura Tipica di Risoluzione

Prima di iniziare a scrivere i cicli, definiamo sempre l'altezza (numero di righe) e la larghezza (numero di colonne) dell'immagine. Questo ci permette di impostare i `range` in modo dinamico e pulito.

Ecco lo schema universale che userai in tutti questi esercizi:

```python
# 1. Definizione delle dimensioni
altezza = len(immagine_test)
larghezza = len(immagine_test[0])

# 2. Doppio ciclo for annidato con range()
for riga in range(altezza):
    for colonna in range(larghezza):
        
        # Seleziono il pixel corrente
        pixel = immagine_test[riga][colonna]
        
        # 3. Controllo con l'istruzione IF
        if pixel[0] == 255:  # Esempio: controllo sul canale Rosso (R)
            # Logica dell'esercizio (es. stampare o incrementare un contatore)
            print(f"Trovato a riga {riga}, colonna {colonna}")
```
