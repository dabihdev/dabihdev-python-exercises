# 🔤 Esercizi sui Metodi delle Stringhe

Benvenuto in questo Jupyter Notebook focalizzato sui metodi di manipolazione delle stringhe in Python. In Python, le stringhe sono **immutabili**, il che significa che qualsiasi metodo applicato non modifica la stringa originale, ma restituisce un nuovo valore modificato che puoi stampare o salvare in una nuova variabile.

### 🛠️ Sintassi e Metodi Principali da Ricordare

Ecco una panoramica dei metodi che utilizzerai negli esercizi:
* `.upper()`: Trasforma tutti i caratteri in MAIUSCOLO.
* `.lower()`: Trasforma tutti i caratteri in minuscolo.
* `.strip()`: Rimuove gli spazi bianchi (o caratteri specifici) all'inizio e alla fine della stringa.
* `.split(separatore)`: Divide una stringa in una lista di sotto-stringhe in base a un separatore.
* `.title()`: Converte la prima lettera di *ogni parola* in maiuscolo e il resto in minuscolo.
* `.capitalize()`: Trasforma in maiuscolo *solo la primissima lettera* dell'intera stringa.
* `.replace(vecchio, nuovo)`: Sostituisce tutte le occorrenze di una sotto-stringa con un'altra.

### 🔗 Pattern Avanzato: Concatenazione dei Metodi (Method Chaining)

Nella seconda parte del notebook incontrerai esercizi che richiedono la **combinazione di più metodi**. In Python puoi concatenare i metodi uno dopo l'altro sulla stessa riga, poiché ogni metodo restituisce a sua volta una stringa.
* *Esempio:* `testo.strip().upper()` prima rimuove gli spazi ai bordi e successivamente converte il risultato pulito in maiuscolo. L'esecuzione avviene sempre da sinistra verso destra.
