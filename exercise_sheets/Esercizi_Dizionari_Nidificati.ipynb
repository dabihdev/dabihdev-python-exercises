## 1. Introduzione Teorica: Cosa sono e come si navigano?

Un **dizionario nidificato** è semplicemente un dizionario che contiene altri dizionari (o liste di dizionari) come valori associati alle sue chiavi. Questo tipo di struttura è fondamentale nella programmazione, poiché permette di rappresentare dati complessi e gerarchici, molto simili al formato JSON utilizzato nel web (es. dati di un'azienda, profili utente, cataloghi prodotti).

### Accesso Diretto
Per accedere a un valore in un dizionario semplice usiamo una singola chiave: `studente['nome']`.
In un dizionario nidificato, si concatenano le parentesi quadre per scendere di livello nella gerarchia:
```python
# Se il dizionario è: scuola = {'classe_1A': {'studente1': {'nome': 'Marco'}}}
nome_studente = scuola['classe_1A']['studente1']['nome']
```
Se a un certo livello incontri una **lista**, dovrai usare l'indice numerico prima di poter accedere nuovamente alle chiavi del dizionario interno:
```python
# Se abbiamo una lista di dizionari: azienda = {'sedi': [{'nome': 'Milano'}, {'nome': 'Roma'}]}
prima_sede = azienda['sedi'][0]['nome'] # Accede a 'Milano'
```

### Navigazione e Filtro con Loop (`for`) e Condizioni (`if`)
Quando non conosciamo l'indice esatto o vogliamo analizzare tutti gli elementi, usiamo i cicli combinati con i metodi dei dizionari:
- `.keys()`: per scorrere le chiavi.
- `.values()`: per scorrere i valori.
- `.items()`: per scorrere contemporaneamente chiavi e valori.

**Pattern tipico di ricerca/filtro:**
```python
for chiave_sede, dati_sede in azienda.items():
    for impiegato in dati_sede['impiegati']:
        if impiegato['stipendio'] > 3000:
            print(impiegato['nome'])
```
