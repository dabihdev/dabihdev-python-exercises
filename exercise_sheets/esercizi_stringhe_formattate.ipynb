## 📘 Guida Rapida alle F-Strings

Le **f-strings** (Formatted String Literals) permettono di inserire variabili ed espressioni direttamente all'interno delle stringhe in modo semplice, veloce e pulito.

### 1. Sintassi Fondamentale
Basta far precedere la stringa dalla lettera `f` o `F` e racchiudere le variabili o espressioni tra parentesi graffe `{}`:
```python
nome = "Annamaria"
print(f"Ciao {nome}!")  # Output: Ciao Annamaria!
```

### 2. Operazioni ed Espressioni Dirette
Dentro le graffe puoi inserire vero e proprio codice Python (calcoli matematici, chiamate a funzioni o metodi):
- **Matematica:** `f"Totale: {10 + 20}"` -> `Totale: 30`
- **Metodi:** `f"Ciao {utente.upper()}"` -> Trasforma il testo in maiuscolo.
- **Funzioni:** `f"Elementi: {len(lista)}"` -> Restituisce la lunghezza della lista.

### 3. Formattazione Avanzata
- **Arrotondare i decimali:** Usa `:.Nf` dove *N* è il numero di cifre desiderate dopo la virgola.
  `f"{3.14159:.2f}"` -> `3.14`
- **Separatore delle migliaia:** Usa `:,` per formattare automaticamente i grandi numeri.
  `f"{59000000:,}"` -> `59,000,000`
- **Allineamento e Riempimento (Padding):** 
  - Centrato con caratteri di riempimento: `f"{'MENU':-^20}"` -> `--------MENU--------` (20 caratteri totali)
