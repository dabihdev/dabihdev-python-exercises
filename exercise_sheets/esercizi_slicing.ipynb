### 📌 Breve Ripasso della Sintassi dello Slicing
La sintassi generale per estrarre porzioni di sequenze è: **`oggetto[inizio:fine:passo]`**
* **`inizio`**: L'indice da cui partire (**incluso**). Se omesso, parte dall'inizio (indice `0`).
* **`fine`**: L'indice a cui fermarsi (**escluso**). Se omesso, arriva fino alla fine dell'oggetto.
* **`passo`**: L'intervallo di quanti elementi saltare alla volta. Se è negativo (es. `-1`), scorre la sequenza al contrario.
