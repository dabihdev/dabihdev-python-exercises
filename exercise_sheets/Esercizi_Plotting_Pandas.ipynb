# Introduzione al Data Plotting con Pandas
Questo notebook è pensato per guidarti passo dopo passo nell'uso delle funzionalità di visualizzazione integrate in **Pandas**.
La libreria Pandas mette a disposizione il metodo `.plot()`, che si appoggia internamente a **Matplotlib**, rendendo la creazione di grafici rapida e intuitiva direttamente a partire da DataFrame e Serie.


## 1. Concetti Principali e Sintassi di Base
Il metodo principale per creare grafici in Pandas è `df.plot()`. I tipi di grafici più comuni possono essere specificati tramite l'argomento `kind` o chiamando direttamente il sotto-metodo specifico:
- `'line'` : Grafico a linee (predefinito) -> `df.plot.line()`
- `'bar'` o `'barh'` : Grafico a barre verticali o orizzontali -> `df.plot.bar()`
- `'hist'` : Istogramma -> `df.plot.hist()`
- `'box'` : Box Plot per la distribuzione statica -> `df.plot.box()`
- `'scatter'` : Grafico a dispersione (richiede `x` e `y`) -> `df.plot.scatter()`
- `'area'` : Grafico ad area -> `df.plot.area()`

### Personalizzazioni Comuni:
- `title`: Stringa per il titolo del grafico.
- `figsize`: Tupla `(larghezza, altezza)` in pollici (es. `figsize=(10, 6)`).
- `color`: Colore delle linee o delle barre.
- `grid`: Booleano (`True`/`False`) per mostrare la griglia di sfondo.
- `xlabel` / `ylabel`: Etichette per l'asse X e l'asse Y (introdotte nelle versioni recenti di Pandas, o modificabili tramite matplotlib).


### Esempio Pratico con Dati Sintetici
Vediamo come creare un semplice DataFrame e farne il plotting immediato.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Per visualizzare i grafici direttamente nel notebook
%matplotlib inline

# Creazione di dati fittizi
date_range = pd.date_range(start='2026-01-01', periods=100, freq='D')
dati_esempio = pd.DataFrame({
    'Vendite': np.random.randint(50, 200, size=100).cumsum(),
    'Profitto': np.random.randint(10, 50, size=100).cumsum()
}, index=date_range)

print('Prime righe del dataset di esempio:')
dati_esempio.head()
```

#### Esempio 1: Grafico a linee predefinito con personalizzazioni
Plottiamo l'andamento di Vendite e Profitto nel tempo.


```python
dati_esempio.plot(figsize=(10, 5), title='Andamento delle Vendite e del Profitto nel 2026', grid=True)
plt.ylabel('Valore in EUR')
plt.xlabel('Data')
plt.show()
```

#### Esempio 2: Istogramma e Scatter Plot
Vediamo come mostrare la distribuzione delle variazioni giornaliere del profitto o una relazione tra due variabili.


```python
# Calcoliamo la variazione giornaliera del profitto
dati_esempio['Variazione_Profitto'] = dati_esempio['Profitto'].diff()

# Istogramma della variazione del profitto
dati_esempio['Variazione_Profitto'].plot(kind='hist', bins=15, color='skyblue', edgecolor='black', figsize=(8, 4), title='Distribuzione Variazione Profitto')
plt.xlabel('Incremento')
plt.show()
```
