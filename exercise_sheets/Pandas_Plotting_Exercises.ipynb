# Introduction to Data Plotting with Pandas
This notebook is designed to guide you step-by-step through the built-in visualization capabilities of **Pandas**.
The Pandas library provides the `.plot()` method, which internally relies on **Matplotlib**, making it quick and intuitive to create charts directly from DataFrames and Series.


## 1. Main Concepts and Basic Syntax
The main method for creating plots in Pandas is `df.plot()`. The most common plot types can be specified using the `kind` argument or by directly calling the specific sub-method:
- `'line'` : Line plot (default) -> `df.plot.line()`
- `'bar'` or `'barh'` : Vertical or horizontal bar chart -> `df.plot.bar()`
- `'hist'` : Histogram -> `df.plot.hist()`
- `'box'` : Box Plot for statistical distribution -> `df.plot.box()`
- `'scatter'` : Scatter plot (requires `x` and `y`) -> `df.plot.scatter()`
- `'area'` : Area plot -> `df.plot.area()`

### Common Customizations:
- `title`: String for the plot title.
- `figsize`: A tuple `(width, height)` in inches (e.g., `figsize=(10, 6)`).
- `color`: Color of lines or bars.
- `grid`: Boolean (`True`/`False`) to show the background grid.
- `xlabel` / `ylabel`: Labels for the X-axis and Y-axis.


### Practical Example with Synthetic Data
Let's see how to create a simple DataFrame and plot it immediately.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# To display plots directly in the notebook
%matplotlib inline

# Creating dummy data
date_range = pd.date_range(start='2026-01-01', periods=100, freq='D')
sample_data = pd.DataFrame({
    'Sales': np.random.randint(50, 200, size=100).cumsum(),
    'Profit': np.random.randint(10, 50, size=100).cumsum()
}, index=date_range)

print('First few rows of the sample dataset:')
sample_data.head()
```

#### Example 1: Default Line Plot with Customizations
Let's plot the trend of Sales and Profit over time.


```python
sample_data.plot(figsize=(10, 5), title='Sales and Profit Trends in 2026', grid=True)
plt.ylabel('Value in EUR')
plt.xlabel('Date')
plt.show()
```

#### Example 2: Histogram and Scatter Plot
Let's see how to show the distribution of daily profit variations or a relationship between two variables.


```python
# Calculate the daily profit change
sample_data['Profit_Change'] = sample_data['Profit'].diff()

# Histogram of profit change
sample_data['Profit_Change'].plot(kind='hist', bins=15, color='skyblue', edgecolor='black', figsize=(8, 4), title='Profit Change Distribution')
plt.xlabel('Increase')
plt.show()
```
