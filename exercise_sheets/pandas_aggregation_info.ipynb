## Quick Guide

### 1. Data Overview Methods

When working with a new dataset, the first step is often to understand its structure, shape, data types, and basic statistics. Pandas provides two essential methods for this:

#### A. `df.info()`
The `.info()` method prints a concise summary of a DataFrame, including:
- The total number of rows (RangeIndex)
- The columns and their data types (dtypes)
- The number of non-null values per column (helping identify missing data)
- Memory usage

**Example:**
```python
import pandas as pd

# Load the dataset
df = pd.read_csv('london_weather.csv', sep=';')

# Get summary information
df.info()
```

#### B. `df.describe()`
The `.describe()` method generates descriptive statistics that summarize the central tendency, dispersion, and shape of a dataset's distribution, excluding `NaN` values. For numeric columns, it provides:
- `count`: Number of non-null observations
- `mean`: The arithmetic average
- `std`: Standard deviation
- `min`: Minimum value
- `25%`, `50%`, `75%`: Percentiles / Quartiles
- `max`: Maximum value

**Example:**
```python
# General statistical summary
df.describe()

# Statistical summary for a specific column
df['max_temp'].describe()

# Custom percentiles
df['max_temp'].describe(percentiles=[0.1, 0.5, 0.9])
```

---

### 2. Basic Aggregation Functions

Aggregation functions summarize a whole series or column into a single statistical value. By default, pandas ignores missing values (`NaN`) when computing these aggregations.

#### A. `.mean()`
Computes the arithmetic mean of the values:
$$\mu = \frac{1}{n} \sum_{i=1}^n x_i$$

```python
df['mean_temp'].mean()
```

#### B. `.min()` and `.max()`
Find the minimum and maximum values in a column:
```python
df['min_temp'].min()
df['max_temp'].max()
```

#### C. `.sum()`
Calculates the sum of all values in a column:
```python
df['precipitation'].sum()
```

#### D. Applying to Multiple Columns
You can call these methods directly on a subset of columns (a DataFrame slice) to get the aggregation for each column:
```python
df[['min_temp', 'max_temp', 'mean_temp']].mean()
```

---

### 3. The `.agg()` Method

The `.agg()` (or `.aggregate()`) method is incredibly versatile. It allows you to:
1. Apply multiple aggregation functions to a single column.
2. Apply different aggregation functions to different columns simultaneously.

#### A. Multiple functions on a single column
Pass a list of function names as strings:
```python
df['sunshine'].agg(['min', 'max', 'mean'])
```

#### B. Different functions on different columns
Pass a dictionary where the keys are column names and the values are lists of functions or a single function:
```python
df.agg({
    'precipitation': 'sum',
    'mean_temp': ['min', 'max', 'mean']
})
```
