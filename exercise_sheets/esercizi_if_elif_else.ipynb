## 🛑 Soluzioni per il Controllo

Usa queste soluzioni solo per verificare il tuo lavoro se rimani bloccato!

```python
# Esercizio 1
if numero > 10:
    print("Il numero è maggiore di 10")

    print("Il numero è 10 o minore")

# Esercizio 2

    print("Sei maggiorenne")
else:
    print("Sei minorenne")

    print("Il numero è negativo")

# Esercizio 6

    print("Fa freddo")
elif gradi < 28:
    print("La temperatura è piacevole")

    print("Fa caldissimo")

# Esercizio 7

    print("Il numero è zero")
```
