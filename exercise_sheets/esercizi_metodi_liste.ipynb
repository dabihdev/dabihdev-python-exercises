# Esercizi sui Metodi delle Liste in Python

Questo notebook contiene 20 esercizi semplici e ripetitivi pensati per automatizzare l'uso dei principali metodi delle liste in Python. Ogni esercizio è strutturato con una breve istruzione seguita da una cella di codice vuota in cui completare l'esercizio.

### Metodi trattati:
- `.append()` - Aggiunge un elemento alla fine
- `.insert()` - Inserisce un elemento a un indice specifico
- `.extend()` - Estende la lista con elementi di un'altra lista
- `.remove()` - Rimuove la prima occorrenza di un valore
- `.pop()` - Rimuove e restituisce un elemento a un dato indice
- `.count()` - Conta le occorrenze di un valore
- `.index()` - Trova l'indice di un valore
- `.sort()` - Ordina la lista
- `.reverse()` - Inverte l'ordine degli elementi
- `.clear()` - Svuota la lista
