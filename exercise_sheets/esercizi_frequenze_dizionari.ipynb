# Il Pattern del Conteggio con i Dizionari

I dizionari in Python sono lo strumento ideale per calcolare le frequenze di elementi (caratteri, parole, numeri, ecc.) grazie alla loro struttura **chiave-valore**. Questo approccio algoritmico può essere chiamato **Pattern del Conteggio**.

### Come funziona la logica di base?
1. Si inizializza un **dizionario vuoto**.
2. Si scorre la collezione (una stringa, una lista, ecc.) elemento per elemento usando un ciclo `for`.
3. Per ogni elemento, si verifica se è già presente tra le chiavi del dizionario:
   * Se **è già presente**, si incrementa il suo valore (il contatore) di 1.
   * Se **non è presente**, significa che lo stiamo vedendo per la prima volta: lo si aggiunge come nuova chiave con un valore iniziale di 1.

### L'approccio classico (if/else):
```python
conteggio = {}
for elemento in collezione:
    if elemento in conteggio:
        conteggio[elemento] += 1
    else:
        conteggio[elemento] = 1
```

### L'approccio compatto ed elegante (con il metodo `.get()`):
Il metodo `.get(chiave, valore_default)` cerca la chiave nel dizionario: se la trova restituisce il suo valore corrente, altrimenti restituisce il `valore_default` (in questo caso `0`). Questo trucco permette di condensare la verifica in una singola riga di codice:
```python
conteggio = {}
for elemento in collezione:
    conteggio[elemento] = conteggio.get(elemento, 0) + 1
```
