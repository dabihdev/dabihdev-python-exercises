### 📌 Breve Ripasso della Sintassi

Le condizioni permettono al programma di prendere decisioni ed eseguire blocchi di codice diversi a seconda che una condizione sia vera (`True`) o falsa (`False`).

```python
if condizione_1:
    # Eseguito se condizione_1 è vera
elif condizione_2:
    # Eseguito se condizione_1 è falsa E condizione_2 è vera
else:
    # Eseguito se tutte le condizioni precedenti sono false
```

*Ricorda l'importanza dell'**indentazione** (i 4 spazi vuoti all'inizio della riga) per definire cosa sta dentro al blocco condizionale.*
