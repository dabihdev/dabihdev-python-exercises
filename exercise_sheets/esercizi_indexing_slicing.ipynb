### 📌 Breve Ripasso della Sintassi

#### 1. Indexing (Indicizzazione)
L'indexing permette di accedere a un singolo elemento di una stringa o di una lista tramite la sua posizione. In Python, gli indici partono sempre da **0**.
* **Indici positivi:** Partono da `0` (primo elemento) e vanno verso destra (`1`, `2`, ecc.).
* **Indici negativi:** Partono da `-1` (ultimo elemento) e vanno verso sinistra (`-2`, `-3`, ecc.).
* **Liste Annidate:** Per accedere a elementi dentro liste contenute in altre liste, si concatenano le parentesi quadre: `matrice[riga][colonna]`.

#### 2. Slicing (Affettamento)
Lo slicing permette di estrarre una parte (una sottosequenza) di una stringa o di una lista.
La sintassi generale è: **`oggetto[inizio:fine:passo]`**
* **`inizio`:** L'indice da cui partire (**incluso**). Se omesso, parte da `0`.
* **`fine`:** L'indice a cui fermarsi (**escluso**). Se omesso, arriva fino all'ultimo elemento.
* **`passo`:** L'intervallo di quanti elementi saltare (opzionale). Se omesso, è `1`. Se impostato a `-1`, scorre la sequenza al contrario.
