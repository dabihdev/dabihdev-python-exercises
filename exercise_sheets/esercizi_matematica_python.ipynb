# Introduzione alle Operazioni Matematiche in Python

Python è uno strumento eccezionale per eseguire calcoli matematici, dalle operazioni aritmetiche elementari fino ad analisi statistiche e ingegneristiche complesse. Di seguito sono elencati gli operatori matematici di base pronti all'uso:

* **Addizione (`+`)**: Somma due valori. (Es: `5 + 3` restituisce `8`)
* **Sottrazione (`-`)**: Sottrae il secondo valore dal primo. (Es: `10 - 4` restituisce `6`)
* **Moltiplicazione (`*`)**: Moltiplica due valori. (Es: `4 * 3` restituisce `12`)
* **Divisione (`/`)**: Divide il primo valore per il secondo e restituisce sempre un numero decimale (`float`). (Es: `7 / 2` restituisce `3.5`)
* **Divisione Intera (`//`)**: Divide due numeri e arrotonda il risultato per difetto al numero intero inferiore. (Es: `7 // 2` restituisce `3`)
* **Modulo (`%`)**: Restituisce il resto della divisione intera. (Es: `7 % 2` restituisce `1`)
* **Esponenziazione (`**`)**: Eleva un numero (base) alla potenza di un altro (esponente). (Es: `2 ** 3` restituisce `8`)

### Precedenza degli Operatori
Python segue le regole matematiche standard per la precedenza delle operazioni (regola PEMDAS: Parentesi, Esponenti, Moltiplicazione/Divisione, Addizione/Sottrazione). È possibile utilizzare le parentesi tonde `()` per raggruppare le espressioni e forzare l'ordine del calcolo desiderato.
