# Data Resampling

### 1. Downsampling (summarizing data)

```python
import pandas as pd
import numpy as np

TIMERANGE_DAYS = 18

# Create a range of 15 days (DatetimeIndex)
dates = pd.date_range('2026-01-01', periods=TIMERANGE_DAYS, freq='D')

# Create a data series with that index
data = pd.Series(np.arange(TIMERANGE_DAYS), index=dates)

# Downsample to Weekly frequency ('W'), taking the mean
weekly_data = data.resample('W').mean()

print("Original Daily Data:")
print(data)
print("\nDownsampled Weekly Mean:")
print(weekly_data)
```

### 2. Upsampling (expanding data)

```python
# Create 3 months of data
monthly_dates = pd.date_range('2026-01-01', periods=3, freq='MS') # MS = Month Start
monthly_series = pd.Series([100, 200, 300], index=monthly_dates)

# Upsample to Daily ('D')
# .ffill() propagates the last valid observation forward to next valid
upsampled_ffill = monthly_series.resample('D').ffill()

# .interpolate() fills gaps linearly
upsampled_interp = monthly_series.resample('D').interpolate(method='linear')

print("Original Monthly Data:")
print(monthly_series)
print("\nUpsampled (Daily) with Forward Fill (First 5):")
print(upsampled_ffill.head())
print("\nUpsampled (Daily) with Linear Interpolation (First 5):")
print(upsampled_interp.head())
```
